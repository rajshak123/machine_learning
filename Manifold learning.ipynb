```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Manifold Learning

```python
from sklearn.datasets import make_s_curve
x,y=make_s_curve(n_samples=1000)
from mpl_toolkits.mplot3d import Axes3D
ax=plt.axes(projection='3d')
ax.scatter3D(x[:,0],x[:,1],x[:,2],c=y)
ax.view_init(10,-60)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2).fit(x)
x_pca=pca.transform(x)
plt.scatter(x_pca[:,0],x_pca[:,1],c=y)
```

#### PCA is linear and you can see that a lot of info was lost and it was unable to recover a lot of info that was embedded in 3d space

#### Manifold is non-linear

```python
from sklearn.manifold import Isomap
iso=Isomap(n_neighbors=20)
iso.fit(x)
x_iso=iso.transform(x)
plt.scatter(x_iso[:,0],x[:,1],c=y)
```

```python
from sklearn.manifold import TSNE
tsne=TSNE()
tsne.fit(x)
x_tsne=tsne.embedding_
```

```python
x_tsne=tsne.fit_transform(x)
plt.scatter(x_tsne[:,0],x_tsne[:,1],c=y)
```

#### TSNE works very well in preserving local structure but not necessarily the global structure

#### TSNE works very well in identifying the groups in the data and often works very well in explorartary data analysis
